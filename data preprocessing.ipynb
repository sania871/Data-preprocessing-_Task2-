```python
print(df.isnull().sum())
df['math_score'] = df['math_score'].fillna(df['math_score'].mean())
df['science_score'] = df['science_score'].fillna(df['science_score'].mean())
df['attendance'] = df['attendance'].fillna(df['attendance'].mean())
df['gender'] = df['gender'].fillna(df['gender'].mode()[0])
df['age'] = df['age'].replace('twenty', 20).fillna(df['age'].mode()[0])
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
```

```python
def remove_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return series.clip(lower, upper)
df['math_score'] = remove_outliers(df['math_score'])
df['science_score'] = remove_outliers(df['science_score'])
```

```python
df = df.drop_duplicates()
```
